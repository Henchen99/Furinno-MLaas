January

```python
input_file_path = "/content/drive/MyDrive/Furinno MLaaS/Sales Diagnostic (Daily)/2022/January 2022/"
excel_file_list = os.listdir(input_file_path)
```

```python
excel_file_list
```

```python
daterange = pd.date_range(start='1/01/2022', end='31/01/2022')
```

```python
df_Jan_22 = pd.DataFrame()
for excel_files, date in zip(excel_file_list, daterange):
     if excel_files.endswith(".csv"):
        df1 = pd.read_csv(input_file_path + excel_files, header=None, skiprows=1) # Turn header into standard row and remove first 2 rows
        df1['date'] = date
        df_Jan_22 = df_Jan_22.append(df1,ignore_index=True)
```
